## Overview

This notebook implements the full computational pipeline described in Sections 2.1–2.12 of the manuscript:

1. **PAW chemistry** — first-order RONS decay (H₂O₂, NO₂⁻, NO₃⁻, ONOO⁻) → stress index *S(t)* (§2.2)
2. **Redox-partitioning controller** — bell-shaped *r*(*S*) that couples ethanol/formate exchange flux ratio (§2.3, central contribution)
3. **Cell death + ATPM burden** — Hill-form kinetics + linear stress-cost augmentation (§2.4)
4. **GSMM + chassis** — iJO1366 / iML1515 × {WT, KO11, <span style="color:#C00000"><b>LY160</b></span>} (§2.5)
5. **dFBA integration** — 10-h batch with Δt = 0.25 h (§2.6)
6. **Analyses 3.1–3.8** — baseline → dose-response → mechanistic decomposition → scenarios → carbon yield → Monte Carlo (N = 1,200; <span style="color:#C00000"><b>EDITED: 200 per chassis x 3 chassis x 2 GSMM</b></span>) → Morris → literature cross-validation

## 3. Model parameters (Table 1 of manuscript)

All parameters centralized in a single class for reproducibility. Monte Carlo ranges as inline comments.

```python
class P:
    """Central parameter container. Values match Table 1 of the manuscript."""

    # --- PAW chemistry (§2.2; Lukes 2014, Tarabova 2018) ---
    H2O2_0     = 5.0    # mM at alpha=1
    NO2_0      = 2.5
    NO3_0      = 1.0
    ONOO_0     = 0.5

    k_H2O2     = 0.277  # h^-1 (t_half = 2.5 h)
    k_NO2      = 0.150
    k_NO3      = 0.020
    k_ONOO     = 1.500

    w_H2O2     = 1.0    # Stress-equivalent weights (Imlay 2008)
    w_NO2      = 0.4
    w_NO3      = 0.05
    w_ONOO     = 5.0

    # --- Cell death + ATPM (§2.4) ---
    kd_max     = 4.0    # h^-1                 MC: [2.5, 6.0]
    S_half_d   = 14.0   # mM-eq
    n_d        = 4
    beta_ATPM  = 0.10   # mmol gDW^-1 h^-1 (mM-eq)^-1  MC: [0.06, 0.14]

    # --- Redox-partitioning controller (§2.3 — central contribution) ---
    r_base     = 0.5    # baseline ethanol/formate ratio (non-binding at S=0)
    r_max      = 10.0   # maximum forced ratio   MC: [6, 14]
    Ka_r       = 1.8    # mM-eq (activation)      MC: [0.8, 3.0]
    n_a        = 2
    Ki_r       = 13.0   # mM-eq (damping)         MC: [9, 16]
    n_i        = 4

    # --- PFL inhibition (Knappe & Sawers 1990; Wagner 1992) ---
    PFL_inh_K  = 0.8    # mM-eq                   MC: [0.4, 1.5]
    PFL_inh_n  = 3

    # --- NADH detoxification burden ---
    gamma_NADH = 0.02   # mmol gDW^-1 h^-1 (mM-eq)^-1  MC: [0.01, 0.04]

    # --- Engineered chassis ---
    ADH_eng    = -40.0  # ALCD2x lower bound for KO11/LY160 (pdc-adhB-cat cassette)   # === EDITED ===

    # --- Integration (§2.6) ---
    dt         = 0.25   # h
    T_batch    = 10.0   # h (40 steps)

    # --- Initial conditions ---
    glc_0      = 55.0   # mM
    X_0        = 0.05   # gDW/L
    glc_uptake_max = 10.0  # mmol gDW^-1 h^-1

print("Parameters loaded.")
print(f"  alpha_PAW design range: [0, 1.5]")
print(f"  Batch horizon: {P.T_batch} h, step {P.dt} h ({int(P.T_batch/P.dt)} steps)")
print(f"  Initial: glucose = {P.glc_0} mM, biomass = {P.X_0} gDW/L")
```

## 4. Core simulation functions

Six functions implement the framework:

| Function | Equation | Section |
|----------|----------|---------|
| `paw_species(t, alpha)` | Eq. (1): analytical RONS decay | §2.2 |
| `stress_S(t, alpha)` | Eq. (2): hydrogen-peroxide-equivalent stress index | §2.2 |
| `r_func(S)` | Eq. (3): bell-shaped controller | §2.3 |
| `pfl_inhibition(S)` | Eq. (5): PFL multiplicative reduction | §2.3 |
| `kd_func(S)` | Eq. (6): Hill-form cell death kernel | §2.4 |
| `atpm_burden(S)` | Eq. (7): ATPM augmentation | §2.4 |

```python
def paw_species(t, alpha):
    """Eq. (1): Analytical first-order decay of four RONS species."""
    H2O2 = alpha * P.H2O2_0 * np.exp(-P.k_H2O2 * t)
    NO2  = alpha * P.NO2_0  * np.exp(-P.k_NO2  * t)
    NO3  = alpha * P.NO3_0  * np.exp(-P.k_NO3  * t)
    ONOO = alpha * P.ONOO_0 * np.exp(-P.k_ONOO * t)
    return H2O2, NO2, NO3, ONOO

def stress_S(t, alpha):
    """Eq. (2): Aggregate hydrogen-peroxide-equivalent stress index."""
    H2O2, NO2, NO3, ONOO = paw_species(t, alpha)
    return P.w_H2O2*H2O2 + P.w_NO2*NO2 + P.w_NO3*NO3 + P.w_ONOO*ONOO

def r_func(S):
    """Eq. (3): Bell-shaped redox-partitioning controller r(S)."""
    activation = (S**P.n_a) / (P.Ka_r**P.n_a + S**P.n_a)
    damping = 1.0 / (1.0 + (S / P.Ki_r)**P.n_i)
    return P.r_base + (P.r_max - P.r_base) * activation * damping

def pfl_inhibition(S):
    """Eq. (5): Multiplicative reduction on PFL upper bound."""
    return 1.0 / (1.0 + (S / P.PFL_inh_K)**P.PFL_inh_n)

def kd_func(S):
    """Eq. (6): Hill-form specific cell death rate."""
    return P.kd_max * (S**P.n_d) / (P.S_half_d**P.n_d + S**P.n_d)

def atpm_burden(S):
    """Eq. (7): Linear augmentation of ATP maintenance flux."""
    return P.beta_ATPM * S + P.gamma_NADH * S

print("Function sanity checks at S = 0, 2, 6, 14 mM-eq:")
print(f"{'S':>6} {'r(S)':>8} {'PFL_inh':>10} {'k_d':>8} {'ATPM':>8}")
for S in [0, 2, 6, 14]:
    print(f"{S:>6.1f} {r_func(S):>8.3f} {pfl_inhibition(S):>10.3f} {kd_func(S):>8.3f} {atpm_burden(S):>8.3f}")
```

## 5. Chassis configuration (§2.5)

| Chassis | Genotype | Implementation |
|---------|----------|----------------|
| **WT**   | Wild-type *E. coli* K-12 | Default reconstruction |
| **KO11** | Δ*frdABCD*, Δ*ldhA*, *pdc-adhB-cat*⁺ | FRD2=0, FRD3=0, LDH_D=0, ALCD2x lb = −40 |
| <span style="color:#C00000"><b>**LY160 proxy**</b></span> | <span style="color:#C00000"><b>KO11 + Δ*ackA* (per Yomano 2008 [14])</b></span> | KO11 + ACKr=0 |

```python
def get_biomass_rxn(model):
    for r in model.reactions:
        if 'BIOMASS' in r.id.upper() and 'core' in r.id.lower():
            return r.id
    for r in model.reactions:
        if 'BIOMASS' in r.id.upper():
            return r.id
    raise ValueError("No biomass reaction found.")

def configure_chassis(model, chassis):
    m = model.copy()
    bio_rxn = get_biomass_rxn(m)
    m.objective = bio_rxn

    m.reactions.EX_o2_e.lower_bound = 0.0
    m.reactions.EX_glc__D_e.lower_bound = -P.glc_uptake_max

    if chassis == 'WT':
        pass
    elif chassis == 'KO11':
        for rid in ['FRD2', 'FRD3']:
            if rid in m.reactions:
                m.reactions.get_by_id(rid).bounds = (0, 0)
        if 'LDH_D' in m.reactions:
            m.reactions.LDH_D.bounds = (0, 0)
        m.reactions.ALCD2x.lower_bound = P.ADH_eng
    elif chassis == 'LY160':   # === EDITED: renamed from 'LY01' to match manuscript ===
        for rid in ['FRD2', 'FRD3']:
            if rid in m.reactions:
                m.reactions.get_by_id(rid).bounds = (0, 0)
        if 'LDH_D' in m.reactions:
            m.reactions.LDH_D.bounds = (0, 0)
        m.reactions.ALCD2x.lower_bound = P.ADH_eng
        if 'ACKr' in m.reactions:
            m.reactions.ACKr.bounds = (0, 0)
    else:
        raise ValueError(f"Unknown chassis: {chassis}")

    coupling_constr = m.problem.Constraint(
        m.reactions.EX_etoh_e.flux_expression - P.r_base * m.reactions.EX_for_e.flux_expression,
        lb=0, ub=None, name='redox_coupling'
    )
    m.add_cons_vars([coupling_constr])
    return m, bio_rxn

# Verify on iJO1366
for chassis in ['WT', 'KO11', 'LY160']:   # === EDITED ===
    m_test, bio = configure_chassis(m_J, chassis)
    sol = m_test.optimize()
    print(f"{chassis:6s}: mu = {sol.objective_value:.4f} h^-1, "
          f"EtOH = {-sol.fluxes['EX_etoh_e']:.2f}, "
          f"Acetate = {-sol.fluxes['EX_ac_e']:.2f} mmol gDW^-1 h^-1")
```

## 6. dFBA integration scheme (§2.6)

Static-optimization approach (Mahadevan et al., 2002). The coupling coefficient is updated via `set_linear_coefficients` on the existing optlang Constraint (8× speed-up vs. remove/re-add).

```python
def update_redox_coupling(m, r_val):
    constr = m.constraints['redox_coupling']
    constr.set_linear_coefficients({
        m.reactions.EX_etoh_e.forward_variable: 1.0,
        m.reactions.EX_etoh_e.reverse_variable: -1.0,
        m.reactions.EX_for_e.forward_variable:  -r_val,
        m.reactions.EX_for_e.reverse_variable:   r_val,
    })

def run_dfba(model, chassis, alpha, T=None, dt=None,
             scenario='full', verbose=False):
    """Run dFBA for one chassis-model combination at given alpha.

    Scenarios: 'full', 'baseline', 'coupling', 'pfl', 'no_alcd'
    """
    if T is None: T = P.T_batch
    if dt is None: dt = P.dt
    n_steps = int(T / dt)

    m, bio_rxn = configure_chassis(model, chassis)
    if scenario == 'no_alcd':
        m.reactions.ALCD2x.bounds = (0, 0)

    pfl_baseline = m.reactions.PFL.upper_bound
    atpm_baseline = m.reactions.ATPM.lower_bound

    t_arr = np.zeros(n_steps + 1)
    X_arr = np.zeros(n_steps + 1)
    S_arr = np.zeros(n_steps + 1)
    glc_arr = np.zeros(n_steps + 1)
    etoh_arr = np.zeros(n_steps + 1)
    for_arr = np.zeros(n_steps + 1)
    ac_arr = np.zeros(n_steps + 1)

    t_arr[0] = 0
    X_arr[0] = P.X_0
    glc_arr[0] = P.glc_0
    S_arr[0] = stress_S(0, alpha)

    for i in range(n_steps):
        t = i * dt
        S_now = stress_S(t, alpha)
        S_arr[i] = S_now

        if scenario == 'baseline':
            r_val = P.r_base
            pfl_factor = 1.0
        elif scenario == 'coupling':
            r_val = r_func(S_now)
            pfl_factor = 1.0
        elif scenario == 'pfl':
            r_val = P.r_base
            pfl_factor = pfl_inhibition(S_now)
        else:
            r_val = r_func(S_now)
            pfl_factor = pfl_inhibition(S_now)

        update_redox_coupling(m, r_val)
        m.reactions.PFL.upper_bound = pfl_baseline * pfl_factor
        m.reactions.ATPM.lower_bound = atpm_baseline + atpm_burden(S_now)

        if glc_arr[i] < 0.1:
            m.reactions.EX_glc__D_e.lower_bound = 0.0
        else:
            m.reactions.EX_glc__D_e.lower_bound = -P.glc_uptake_max

        try:
            sol = m.optimize()
            if sol.status != 'optimal':
                X_arr[i+1] = X_arr[i] * np.exp(-kd_func(S_now) * dt)
                glc_arr[i+1] = glc_arr[i]; etoh_arr[i+1] = etoh_arr[i]
                for_arr[i+1] = for_arr[i]; ac_arr[i+1] = ac_arr[i]
                t_arr[i+1] = t + dt
                continue

            mu = sol.objective_value
            v_glc = sol.fluxes['EX_glc__D_e']
            v_etoh = sol.fluxes['EX_etoh_e']
            v_for = sol.fluxes['EX_for_e']
            v_ac = sol.fluxes['EX_ac_e']

            kd = kd_func(S_now)
            X_arr[i+1] = X_arr[i] * np.exp((mu - kd) * dt)
            glc_arr[i+1] = max(0, glc_arr[i] + v_glc * X_arr[i] * dt)
            etoh_arr[i+1] = etoh_arr[i] + v_etoh * X_arr[i] * dt
            for_arr[i+1] = for_arr[i] + v_for * X_arr[i] * dt
            ac_arr[i+1] = ac_arr[i] + v_ac * X_arr[i] * dt
            t_arr[i+1] = t + dt
        except Exception as e:
            if verbose:
                print(f"  Step {i} failed: {e}")
            X_arr[i+1] = X_arr[i]; glc_arr[i+1] = glc_arr[i]
            etoh_arr[i+1] = etoh_arr[i]; for_arr[i+1] = for_arr[i]
            ac_arr[i+1] = ac_arr[i]; t_arr[i+1] = t + dt

    S_arr[-1] = stress_S(T, alpha)
    glc_consumed = P.glc_0 - glc_arr[-1]
    yield_etoh = etoh_arr[-1] / glc_consumed if glc_consumed > 0 else 0

    return {
        't': t_arr, 'X': X_arr, 'S': S_arr,
        'glc': glc_arr, 'etoh': etoh_arr, 'for': for_arr, 'ac': ac_arr,
        'EtOH_final': etoh_arr[-1], 'For_final': for_arr[-1],
        'Ac_final': ac_arr[-1], 'X_final': X_arr[-1],
        'peak_S': S_arr.max(), 'yield': yield_etoh,
    }

print("dFBA integrator ready.")
```

## 12. Analysis 3.6 — Monte Carlo robustness (§3.6)

Latin hypercube sampling: **N = 300 per chassis × 1,800 simulations + 100 enrichment per condition** in [0.5, 0.7].
Takes ~15-25 min on Colab CPU.

```python
mc_ranges = {
    'alpha':     (0.0, 1.5),
    'r_max':     (6.0, 14.0),
    'kd_max':    (2.5, 6.0),
    'beta_ATPM': (0.06, 0.14),
    'Ka_r':      (0.8, 3.0),
    'Ki_r':      (9.0, 16.0),
    'PFL_inh_K': (0.4, 1.5),
    'gamma_NADH':(0.01, 0.04),
}

def latin_hypercube(n_samples, n_dims, seed=42):
    rng = np.random.default_rng(seed)
    samples = np.zeros((n_samples, n_dims))
    for d in range(n_dims):
        perm = rng.permutation(n_samples)
        samples[:, d] = (perm + rng.uniform(0, 1, n_samples)) / n_samples
    return samples

def mc_run(chassis, model, n=100, n_enrichment=100, seed=42):
    """One chassis-model combination: n base samples + n_enrichment in alpha in [0.5, 0.7].
    Returns a DataFrame with 200 rows (100 + 100 enrichment) per chassis-model combination."""
    n_dims = len(mc_ranges)
    lhs = latin_hypercube(n, n_dims, seed)

    lhs_enrich = latin_hypercube(n_enrichment, n_dims, seed + 1)
    lhs_enrich[:, 0] = (lhs_enrich[:, 0] * (0.7 - 0.5) + 0.5) / 1.5
    lhs_all = np.vstack([lhs, lhs_enrich])

    keys = list(mc_ranges.keys())
    results = []

    for idx, s in enumerate(tqdm(lhs_all, desc=f'MC {chassis}', leave=False)):
        params = {}
        for k, val_norm in zip(keys, s):
            lo, hi = mc_ranges[k]
            params[k] = lo + val_norm * (hi - lo)

        orig = {k: getattr(P, k) for k in keys if k != 'alpha'}
        for k, v in params.items():
            if k != 'alpha':
                setattr(P, k, v)

        try:
            res = run_dfba(model, chassis, params['alpha'])
            row = {'mc_idx': idx, **params,
                   'EtOH': res['EtOH_final'], 'peak_S': res['peak_S'],
                   'X_final': res['X_final'], 'yield': res['yield']}
            results.append(row)
        except Exception:
            pass
        finally:
            for k, v in orig.items():
                setattr(P, k, v)

    return pd.DataFrame(results)

# === EDITED: now run BOTH iJO1366 (variant A) AND iML1515 (variant B), and save under canonical filenames ===
print("Running Monte Carlo (iJO1366 + iML1515, all chassis): 200 samples per chassis-model combination...")
mc_per_chassis = {}
for chassis in ['WT', 'KO11', 'LY160']:
    chunks = []
    for variant, gsmm_name, m in [('A', 'iJO1366', m_J), ('B', 'iML1515', m_M)]:
        df_mc = mc_run(chassis, m, n=100, n_enrichment=100, seed=42)
        df_mc.insert(0, 'chassis', chassis)
        df_mc.insert(1, 'variant', variant)
        chunks.append(df_mc)
    mc_per_chassis[chassis] = pd.concat(chunks, ignore_index=True)

# Save per-chassis files (canonical names used by the manuscript)
mc_per_chassis['WT'].to_csv(os.path.join(OUTPUT_DIR, 'mc_full.csv'), index=False)
mc_per_chassis['KO11'].to_csv(os.path.join(OUTPUT_DIR, 'mc_KO11.csv'), index=False)
mc_per_chassis['LY160'].to_csv(os.path.join(OUTPUT_DIR, 'mc_LY160.csv'), index=False)   # === EDITED: renamed from mc_LY01.csv ===

# Combined view for downstream analyses
df_mc_full = pd.concat(list(mc_per_chassis.values()), ignore_index=True)
print(f"\nTotal Monte Carlo samples: {len(df_mc_full)}   (expected 1,200 = 200 x 3 chassis x 2 GSMM)")
```

```python
# Operating-window statistics
print("Operating-window robustness (alpha in [0.5, 0.7]):")
print("=" * 80)
print(f"{'Chassis':10s} {'N':>5s} {'Mean':>7s} {'SD':>7s} {'CV (%)':>8s} {'P5':>6s} {'P95':>6s} {'Safe%':>7s} {'Viab%':>7s}")
print("-" * 80)
for chassis in ['WT', 'KO11', 'LY160']:   # === EDITED ===
    sub = df_mc_full[(df_mc_full['chassis']==chassis) &
                     (df_mc_full['alpha']>=0.5) &
                     (df_mc_full['alpha']<=0.7)]
    if len(sub) > 5:
        cv = (sub['EtOH'].std() / sub['EtOH'].mean()) * 100
        safe_pct = (sub['peak_S'] < 10.0).mean() * 100
        viab_pct = (sub['X_final'] > 0.1).mean() * 100
        print(f"{chassis:10s} {len(sub):>5d} {sub['EtOH'].mean():>7.2f} "
              f"{sub['EtOH'].std():>7.2f} {cv:>8.2f} "
              f"{sub['EtOH'].quantile(0.05):>6.2f} {sub['EtOH'].quantile(0.95):>6.2f} "
              f"{safe_pct:>7.1f} {viab_pct:>7.1f}")
```

## 13. Analysis 3.7 — Morris elementary-effects screening (§3.7)

```python
def morris_screening(model, chassis, n_traj=6, seed=123):
    keys = list(mc_ranges.keys())
    n_params = len(keys)
    rng = np.random.default_rng(seed)

    elem_effects = {k: [] for k in keys}

    for traj in tqdm(range(n_traj), desc=f'Morris {chassis}', leave=False):
        base_norm = rng.uniform(0, 1, n_params)
        delta = 0.5
        for i, k in enumerate(keys):
            perturbed_norm = base_norm.copy()
            perturbed_norm[i] = min(1.0, perturbed_norm[i] + delta)

            y_base, y_pert = 0, 0
            for params_norm, label in [(base_norm, 'base'), (perturbed_norm, 'pert')]:
                params = {}
                for k2, v_norm in zip(keys, params_norm):
                    lo, hi = mc_ranges[k2]
                    params[k2] = lo + v_norm * (hi - lo)

                orig = {k2: getattr(P, k2) for k2 in keys if k2 != 'alpha'}
                for k2, v in params.items():
                    if k2 != 'alpha':
                        setattr(P, k2, v)

                try:
                    res = run_dfba(model, chassis, params['alpha'])
                    if label == 'base':
                        y_base = res['EtOH_final']
                    else:
                        y_pert = res['EtOH_final']
                except Exception:
                    pass
                finally:
                    for k2, v in orig.items():
                        setattr(P, k2, v)

            ee = (y_pert - y_base) / delta
            elem_effects[k].append(ee)

    results = []
    for k in keys:
        ees = np.array(elem_effects[k])
        mu_star = np.abs(ees).mean()
        sigma = ees.std()
        results.append({'param': k, 'mu_star': mu_star, 'sigma': sigma,
                       'sigma_ratio': sigma/mu_star if mu_star > 0 else 0})

    return pd.DataFrame(results).sort_values('mu_star', ascending=False)

print("Morris screening (WT/iJO1366)...")
df_morris = morris_screening(m_J, 'WT', n_traj=6)
print()
print(df_morris.to_string(index=False, float_format=lambda x: f'{x:.3f}'))

df_morris.to_csv(os.path.join(OUTPUT_DIR, 'morris_sensitivity.csv'), index=False)
```

## 15. Adversarial sensitivity analysis (§3.8,  <span style="color:#C00000"><b>— NEWLY ADDED to reproduce adversarial_sweeps.csv and adversarial_summary.csv</b></span>

Three structural-assumption stress tests, each on a fine α grid for the WT/iJO1366 configuration:

| Scenario | Modification | Tests |
|----------|--------------|-------|
| **nominal** | No change (diagnostic grid) | baseline reference |
| **AS-1** | wONOO 5 → 50 (×10) | RONS-weight assumption |
| **AS-2** | Remove damping factor (Ki → ∞) | bell-shape vs sigmoid |
| **AS-3** | Disable PFL inhibition arm (Eq. 5) | PFL-inhibition necessity |

Note: this diagnostic grid uses a finer α step than the main sweep, and the WT/iJO1366 baseline reads 17.0 mM here vs 18.0 mM in `sweep_v4_combined.csv`. This discrepancy is documented in §3.8 of the manuscript.

```python
# === EDITED: NEW SECTION — runs the four adversarial scenarios and writes adversarial_sweeps.csv + adversarial_summary.csv ===

def run_adversarial(scenario_name, alphas_grid):
    """Run a full dose-response under one adversarial scenario."""
    orig_w_ONOO = P.w_ONOO
    orig_Ki_r = P.Ki_r
    rows = []
    for alpha in tqdm(alphas_grid, desc=f'AS {scenario_name}', leave=False):
        # apply scenario perturbation
        if scenario_name == 'AS-1':           # peroxynitrite weight x10
            P.w_ONOO = orig_w_ONOO * 10
        elif scenario_name == 'AS-2':         # remove damping arm
            P.Ki_r = 1e6
        # AS-3 handled inside run_dfba via scenario kwarg below

        if scenario_name == 'AS-3':
            res = run_dfba(m_J, 'WT', alpha, scenario='coupling')   # coupling only, no PFL inh
        else:
            res = run_dfba(m_J, 'WT', alpha)

        rows.append({
            'alpha': alpha, 'scenario': scenario_name,
            'EtOH': res['EtOH_final'], 'Ac': res['Ac_final'], 'For': res['For_final'],
            'peak_S': res['peak_S'], 'yield': res['yield']
        })

        # restore
        P.w_ONOO = orig_w_ONOO
        P.Ki_r = orig_Ki_r
    return pd.DataFrame(rows)

alphas_fine = np.round(np.arange(0.0, 1.55, 0.05), 2)   # 31 doses, step 0.05
adv_frames = [run_adversarial(name, alphas_fine) for name in ['nominal', 'AS-1', 'AS-2', 'AS-3']]
df_adv = pd.concat(adv_frames, ignore_index=True)

# Summary: peak per scenario
summary_rows = []
for sc in ['nominal', 'AS-1', 'AS-2', 'AS-3']:
    s = df_adv[df_adv['scenario'] == sc].sort_values('EtOH', ascending=False)
    base = df_adv[(df_adv['scenario'] == sc) & (df_adv['alpha'] == 0)]['EtOH'].iloc[0]
    pk = s.iloc[0]
    summary_rows.append({
        'scenario': sc, 'baseline_EtOH': base, 'peak_EtOH': pk['EtOH'],
        'peak_alpha': pk['alpha'], 'fold_change': pk['EtOH'] / base
    })
df_adv_sum = pd.DataFrame(summary_rows)
print(df_adv_sum.to_string(index=False))

df_adv.to_csv(os.path.join(OUTPUT_DIR, 'adversarial_sweeps.csv'), index=False)
df_adv_sum.to_csv(os.path.join(OUTPUT_DIR, 'adversarial_summary.csv'), index=False)
print("\nSaved: adversarial_sweeps.csv, adversarial_summary.csv")
```

## 16. Controller benchmark — A vs B  <span style="color:#C00000"><b>— NEWLY ADDED to reproduce controller_benchmark_sweeps.csv and controller_benchmark_summary.csv</b></span>

| Config | Controller | Enzyme target | GSMM |
|--------|------------|---------------|------|
| **A_ALCD_iJO** | A (single-enzyme Hill activation) | ALCD2x | iJO1366 |
| **A_ALCD_iML** | A (single-enzyme Hill activation) | ALCD2x | iML1515 |
| **A_ADHE_iJO** | A (single-enzyme Hill activation) | ADHEr  | iJO1366 |
| **B_iJO**      | B (redox-partitioning, this work) | — (network ratio) | iJO1366 |
| **B_iML**      | B (redox-partitioning, this work) | — (network ratio) | iML1515 |

Three discriminating metrics: enzyme-choice fragility (A on ALCD2x vs ADHEr), cross-GSMM concordance, by-product realism.



```python
# === EDITED: NEW SECTION — Controller A (single-enzyme Hill) vs Controller B (redox-partitioning, this work) ===

def run_dfba_controller_A(model, enzyme_rxn, alpha, M_max=1.0, K_M=1.8, n_M=2):
    """Controller A: single-enzyme Hill activation. Multiplies the upper bound of `enzyme_rxn`
    by (1 + M_max * S^n / (K_M^n + S^n)). All else identical to run_dfba (WT chassis)."""
    dt = P.dt; n_steps = int(P.T_batch / dt)
    m, bio_rxn = configure_chassis(model, 'WT')
    base_ub = m.reactions.get_by_id(enzyme_rxn).upper_bound
    atpm_baseline = m.reactions.ATPM.lower_bound
    pfl_baseline = m.reactions.PFL.upper_bound

    X = P.X_0; glc = P.glc_0; etoh = 0.0; for_ = 0.0; ac = 0.0; peak_S = 0.0
    for i in range(n_steps):
        t = i * dt
        S = stress_S(t, alpha); peak_S = max(peak_S, S)
        # Controller A: boost the chosen enzyme's UB, keep r at r_base, keep PFL untouched
        m.reactions.get_by_id(enzyme_rxn).upper_bound = base_ub * (1.0 + M_max * S**n_M / (K_M**n_M + S**n_M))
        update_redox_coupling(m, P.r_base)
        m.reactions.PFL.upper_bound = pfl_baseline   # no PFL inhibition arm in Controller A
        m.reactions.ATPM.lower_bound = atpm_baseline + atpm_burden(S)
        m.reactions.EX_glc__D_e.lower_bound = 0.0 if glc < 0.1 else -P.glc_uptake_max

        try:
            sol = m.optimize()
            if sol.status != 'optimal':
                X = X * np.exp(-kd_func(S) * dt); continue
            mu = sol.objective_value
            v_glc = sol.fluxes['EX_glc__D_e']; v_etoh = sol.fluxes['EX_etoh_e']
            v_for = sol.fluxes['EX_for_e']; v_ac = sol.fluxes['EX_ac_e']
            kd = kd_func(S)
            X = X * np.exp((mu - kd) * dt)
            glc = max(0, glc + v_glc * X * dt)
            etoh += v_etoh * X * dt; for_ += v_for * X * dt; ac += v_ac * X * dt
        except Exception:
            pass
    return {'EtOH_final': etoh, 'For_final': for_, 'Ac_final': ac, 'X_final': X, 'peak_S': peak_S}


alphas_bench = np.round(np.arange(0.0, 1.55, 0.05), 2)
bench_rows = []
configs = [
    ('A_ALCD_iJO', 'A', 'ALCD2x', 'iJO1366', m_J),
    ('A_ALCD_iML', 'A', 'ALCD2x', 'iML1515', m_M),
    ('A_ADHE_iJO', 'A', 'ADHEr',  'iJO1366', m_J),
    ('B_iJO',      'B', None,     'iJO1366', m_J),
    ('B_iML',      'B', None,     'iML1515', m_M),
]
for cfg, ctrl, enz, gsmm, m in configs:
    for alpha in tqdm(alphas_bench, desc=f'Bench {cfg}', leave=False):
        if ctrl == 'A':
            res = run_dfba_controller_A(m, enz, alpha)
        else:
            res = run_dfba(m, 'WT', alpha)   # Controller B = this work
        bench_rows.append({
            'alpha': alpha, 'controller': ctrl, 'enzyme': enz or 'ratio',
            'gsmm': gsmm, 'EtOH': res['EtOH_final'], 'Ac': res['Ac_final'],
            'For': res['For_final'], 'peak_S': res['peak_S'], 'config': cfg
        })
df_bench = pd.DataFrame(bench_rows)

# Summary metrics
def fold(cfg):
    s = df_bench[df_bench['config'] == cfg]
    base = s[s['alpha'] == 0]['EtOH'].iloc[0]
    return s['EtOH'].max() / base

fold_A_ALCD_iJO = fold('A_ALCD_iJO')
fold_A_ADHE_iJO = fold('A_ADHE_iJO')
fold_A_ALCD_iML = fold('A_ALCD_iML')
fold_B_iJO = fold('B_iJO')
fold_B_iML = fold('B_iML')

def suppr(cfg, metric, alpha_target=0.60):
    s = df_bench[df_bench['config'] == cfg]
    base = s[s['alpha'] == 0][metric].iloc[0]
    val = s.iloc[(s['alpha'] - alpha_target).abs().argsort().iloc[0]][metric]
    return (1 - val / base) if base > 0 else 0

summary_rows = [
    {'metric': 'Hormetic fold-change (iJO1366)',
     'Controller_A': f'{fold_A_ALCD_iJO:.2f}x',
     'Controller_B': f'{fold_B_iJO:.2f}x'},
    {'metric': 'Enzyme-choice fragility',
     'Controller_A': f'{abs(fold_A_ALCD_iJO - fold_A_ADHE_iJO):.2f}',
     'Controller_B': '0.00 (no free choice)'},
    {'metric': 'Cross-GSMM fold spread',
     'Controller_A': f'{abs(fold_A_ALCD_iJO - fold_A_ALCD_iML):.3f}',
     'Controller_B': f'{abs(fold_B_iJO - fold_B_iML):.3f}'},
    {'metric': 'Acetate suppression',
     'Controller_A': f'{suppr("A_ALCD_iJO", "Ac"):.0%}',
     'Controller_B': f'{suppr("B_iJO", "Ac"):.0%}'},
    {'metric': 'Formate suppression',
     'Controller_A': f'{suppr("A_ALCD_iJO", "For"):.0%}',
     'Controller_B': f'{suppr("B_iJO", "For"):.0%}'},
]
df_bench_sum = pd.DataFrame(summary_rows)
print(df_bench_sum.to_string(index=False))

df_bench.to_csv(os.path.join(OUTPUT_DIR, 'controller_benchmark_sweeps.csv'), index=False)
df_bench_sum.to_csv(os.path.join(OUTPUT_DIR, 'controller_benchmark_summary.csv'), index=False)
print("\nSaved: controller_benchmark_sweeps.csv, controller_benchmark_summary.csv")
```
